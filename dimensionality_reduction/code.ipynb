```python
import numpy as np
from sklearn.datasets import load_wine
from dimensionality_reduction.pca import PCA
```

```python
data = load_wine()
X = data.data
y = data.target

print("Original shape:", X.shape)

pca = PCA(n_components=5)
X_reduced = pca.fit_transform(X)

print("Reduced shape:", X_reduced.shape)

print("\nFirst 5 transformed samples:\n")
print(X_reduced[:5])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_reduced, y, test_size=0.2, random_state=42
)
```

```python
from supervised.linear_models.logistic_regression import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from metrics.classification import accuracy_score, precision_score, recall_score, f1_score
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")



```
