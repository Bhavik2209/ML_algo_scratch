```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

from unsupervised.DBSCAN.DBSCAN import DBSCAN
from metrics.clustering.silhouette_score import ShilhouetteScore


# 1. Create dataset
X, y_true = make_moons(n_samples=300, noise=0.05, random_state=42)

# 2. Apply Scratch DBSCAN
db = DBSCAN(eps=0.3, min_samples=5)
db.fit(X)

labels = db.labels

# 3. Compute Silhouette Score
score = ShilhouetteScore.calculate(X, labels)

print("Silhouette Score:", score)


# 4. Visualize result
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("Scratch DBSCAN Clustering")
plt.show()
```
