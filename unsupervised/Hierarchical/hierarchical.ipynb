```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering

from unsupervised.Hierarchical.hierarchical import HierarchicalClustering
from metrics.clustering.silhouette_score import ShilhouetteScore


# Load dataset
data = load_wine()
X = data.data

# Scale data (important for distance-based clustering)
scaler = StandardScaler()
X = scaler.fit_transform(X)


# ---------------- SCRATCH DIVISIVE ----------------
model_div = HierarchicalClustering(
    n_clusters=3,
    linkage='average',
    method='divisive'
)

model_div.fit(X)
labels_div = model_div.labels

score_div = ShilhouetteScore.calculate(X, labels_div)
print("Scratch Divisive Silhouette Score:", score_div)


model_agg = HierarchicalClustering(
    n_clusters=3,
    linkage='average',
    method='agglomerative'
)

model_agg.fit(X)
labels_agg = model_agg.labels

score_agg = ShilhouetteScore.calculate(X, labels_agg)
print("Scratch Agglomerative Silhouette Score:", score_agg)


sk_model = AgglomerativeClustering(n_clusters=3, linkage='average')
labels_sk = sk_model.fit_predict(X)

score_sk = ShilhouetteScore.calculate(X, labels_sk)
print("Sklearn Agglomerative Silhouette Score:", score_sk)
```
