```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt


from unsupervised.kmeans.kmeans import KMeans
from metrics.clustering.silhouette_score import ShilhouetteScore


X, _ = make_blobs(
    n_samples=300,
    centers=3,
    cluster_std=1.0,
    random_state=42
)

kmeans = KMeans(k=3, max_iterations=100)
kmeans.fit(X)

labels = kmeans.predict(X)


score = ShilhouetteScore.calculate(X, labels)

print("Silhouette Score:", score)

plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(
    kmeans.centroids[:, 0],
    kmeans.centroids[:, 1],
    marker='X',
    s=200
)
plt.title("Scratch KMeans Clustering")
plt.show()
```
